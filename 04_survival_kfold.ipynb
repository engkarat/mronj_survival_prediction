```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
f_path = './All-DATA-MRONJ-Deep-learning_2023-09-18.csv'
df = pd.read_csv(f_path)
# df = df[df['Age'] != '-']
df = df[df['MRONJ'] != 0]
df
```

```python
cols = [
    # 'Age_range', 'MRONJ_related_Disease', 'Underlying_Disease', 'Duration', 'Severity', 'DurationRANGE',
    'Gender', 'Age', 'MRONJ_related_Disease_separate',
    'Underlying_Disease_separate', 'Antiresorptive_drug', 'SteroidDrug',
    'ImmunosuppressiveDrug', 'Calcium', 'Location', 'DentalHistory', 
    'MRONJ', 'Duration', 'Severity'
]
df[cols]
```

# Adjust value of 'Gender' to 0/1 and 'Age' to z-score

```python
df_adj = df[cols].copy()
df_adj['Gender'] = df_adj['Gender'] - 1
df_adj['Age'] = df_adj['Age'].astype('float')
df_adj['Age'] = (df_adj['Age'] - np.mean(df_adj['Age'])) / np.std(df_adj['Age'])
# df_adj['Age_range'] = df_adj['Age_range'] - 1
df_adj
```

```python
cat_col = [
    # 'Location', 'DentalHistory'
    'SteroidDrug', 'ImmunosuppressiveDrug', 'Calcium', 'MRONJ_related_Disease_separate', 'Underlying_Disease_separate', 'Antiresorptive_drug', 
]
df_adj[cat_col]
```

# Convert categorical data to one-hot encoding

```python
cat_feat = pd.get_dummies(df_adj[cat_col].astype('str'))
cat_feat
```

```python
num_feat = df_adj[['Gender', 'Age']]
num_feat
```

# Run K-fold (K=10), train model, and evaluate

```python
from sklearn.linear_model import LogisticRegression
from sklearn.svm import SVC
from sklearn.tree import DecisionTreeClassifier
import sklearn.metrics as sk

lbl_col = 'Duration'
train_data = pd.concat([num_feat, cat_feat], axis=1)
label_data = df_adj[[lbl_col]] - 1
k = 5
auc_roc_list = []

for n in range(5):
    np.random.seed(123 + n)
    train_data_neg = train_data[label_data['Duration']==0]
    label_data_neg = label_data[label_data['Duration']==0]
    idx_neg = list(range(len(train_data_neg)))
    rand_idx_neg = np.random.permutation(idx_neg)

    train_data_pos = train_data[label_data['Duration']==1]
    label_data_pos = label_data[label_data['Duration']==1]
    idx_pos = list(range(len(train_data_pos)))
    rand_idx_pos = np.random.permutation(idx_pos)

    for i in range(k):
        # Negative
        test_size_neg = len(train_data_neg) // k
        sel_idx_neg = rand_idx_neg[test_size_neg*i :test_size_neg*(i+1)]
        te_x_neg = train_data_neg.iloc[sel_idx_neg]
        te_y_neg = label_data_neg.iloc[sel_idx_neg]
        tr_x_neg = train_data_neg.iloc[rand_idx_neg[~np.isin(rand_idx_neg, sel_idx_neg)]]
        tr_y_neg = label_data_neg.iloc[rand_idx_neg[~np.isin(rand_idx_neg, sel_idx_neg)]]

        # Possitive
        test_size_pos = len(train_data_pos) // k
        sel_idx_pos = rand_idx_pos[test_size_pos*i :test_size_pos*(i+1)]
        te_x_pos = train_data_pos.iloc[sel_idx_pos]
        te_y_pos = label_data_pos.iloc[sel_idx_pos]
        tr_x_pos = train_data_pos.iloc[rand_idx_pos[~np.isin(rand_idx_pos, sel_idx_pos)]]
        tr_y_pos = label_data_pos.iloc[rand_idx_pos[~np.isin(rand_idx_pos, sel_idx_pos)]]

        # Merge
        tr_x = pd.concat([tr_x_pos, tr_x_neg])
        te_x = pd.concat([te_x_pos, te_x_neg])
        tr_y = pd.concat([tr_y_pos, tr_y_neg])
        te_y = pd.concat([te_y_pos, te_y_neg])

        # Train model
        model = LogisticRegression(max_iter=1000)
        # model = DecisionTreeClassifier()
        model.fit(tr_x, tr_y[lbl_col])
        pred = model.predict_proba(te_x)

        # Evaluate with AUC-ROC
        auc_roc = sk.roc_auc_score(te_y[lbl_col], pred[:, 1])
        print(f"{i} - AUC-ROC = {auc_roc*100:.2f}%")
        auc_roc_list.append(auc_roc)
    print('=======================')
print(f"Average AUC-ROC = {np.mean(auc_roc_list)*100:.2f}%")
```
