```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
f_path = './All-DATA-MRONJ-Deep-learning_2023-09-18.csv'
df = pd.read_csv(f_path)
# df = df[df['Age'] != '-']
df = df[df['MRONJ'] != 0]
df
```

```python
cols = [
    # 'Age_range', 'MRONJ_related_Disease', 'Underlying_Disease', 'Duration', 'Severity', 'DurationRANGE',
    'Gender', 'Age', 'MRONJ_related_Disease_separate',
    'Underlying_Disease_separate', 'Antiresorptive_drug', 'SteroidDrug',
    'ImmunosuppressiveDrug', 'Calcium', 'Location', 'DentalHistory', 
    'MRONJ', 'Duration', 'Severity'
]
df[cols]
```

# Adjust value of 'Gender' to 0/1 and 'Age' to z-score

```python
df_adj = df[cols].copy()
df_adj['Gender'] = df_adj['Gender'] - 1
df_adj['Age'] = df_adj['Age'].astype('float')
df_adj['Age'] = (df_adj['Age'] - np.mean(df_adj['Age'])) / np.std(df_adj['Age'])
# df_adj['Age_range'] = df_adj['Age_range'] - 1
df_adj
```

```python
cat_col = [
    # 'Location', 'DentalHistory'
    'SteroidDrug', 'ImmunosuppressiveDrug', 'Calcium', 'MRONJ_related_Disease_separate', 'Underlying_Disease_separate', 'Antiresorptive_drug', 
]
df_adj[cat_col]
```

# Convert categorical data to one-hot encoding

```python
cat_feat = pd.get_dummies(df_adj[cat_col].astype('str'))
cat_feat
```

```python
num_feat = df_adj[['Gender', 'Age']]
num_feat
```

## SMOTE oversampling

```python
lbl_col = 'Duration'
train_data = pd.concat([num_feat, cat_feat], axis=1)
label_data = df_adj[[lbl_col]] - 1
```

```python
train_data.shape, label_data.shape
```

```python
np.unique(label_data, return_counts=True)
```

```python
import imblearn
```

```python
oversample = imblearn.over_sampling.SMOTE()
train_data_imb, label_data_imb = oversample.fit_resample(train_data, label_data)
```

```python
train_data_imb.shape, label_data_imb.shape
```

```python
np.unique(label_data_imb, return_counts=True)
```
