```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
f_path = './All-DATA-MRONJ-Deep-learning_2023-09-18.csv'
df = pd.read_csv(f_path)
df
```

```python
age = df['Age'].apply(lambda x: -1 if x == '-' else x).astype('int')
plt.figure(figsize=[10, 6])
plt.hist(age)
plt.xticks(rotation=90)
```

```python
age_clean = age[age > 0]
np.min(age_clean), np.max(age_clean), np.median(age_clean), np.mean(age_clean)
```
